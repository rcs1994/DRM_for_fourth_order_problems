```python
import matplotlib.pyplot as plt
import numpy as np
import pickle as pkl
```

```python
with open("results/loss.pkl",'rb') as pfile:
    hist = pkl.load(pfile)
```

```python
plt.plot(hist, label='loss')
plt.yscale('linear')  # Set y-axis to logarithmic scale, or use 'linear' if you want normal scale
plt.ylim(-2e-1, 1e-1)
plt.legend()
plt.show()
```
